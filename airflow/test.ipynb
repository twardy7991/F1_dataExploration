```python
import fastf1

race = fastf1.get_session(2023, "Bahrain", "Qualifying")
```

```python

race.load(weather=None, messages=None)
```

```python
laps = race.laps
```

```python
laps.sort_values(by="LapTime").head(1)
```

```python
driver_laps = laps[laps["Driver"] == "VER"]
```

```python
driver_laps
```

```python
race.laps.pick_driver("VER").pick_fastest()
```

```python
from pandas.testing import assert_frame_equal 

assert_frame_equal(driver_laps, race.laps.pick_driver("VER"))
```

```python
race.laps.pick_driver("VER").pick_fastest().get_telemetry()
```

```python
assert_frame_equal(laps.sort_values(by="LapTime").head(1),  race.laps.pick_driver("VER").pick_fastest())
```
